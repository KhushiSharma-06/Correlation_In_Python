```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

#read in the data
df=pd.read_csv('movies.csv')
```

```python
#taking a glimpse of the dataset
df.head()
```

```python
# checking for any missing data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Impute with value 0 on columns with null values
for col in df.columns:
    df[col]=df[col].fillna(0)
```

```python
# again checking for null values
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Datatypes of the columns
df.dtypes
```

```python
# sorting the dataframe 
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# dropping the duplicates
df.drop_duplicates()
```

```python
# scatterplot with budget and gross revenue


plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for the film')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Correlation Matrix For Numeric Values
df.corr(method ='pearson',numeric_only=True)
# Corr only works on numeric fields and not fields like company name etc
# There are different types of correlation like pearson (default), kendall, spearman
# numeric_only has a default value false
```

```python
# Correlation Matrix For Numeric Values
df.corr(method ='spearman',numeric_only=True)
```

```python
# Correlation Matrix For Numeric Values
df.corr(method ='kendall',numeric_only=True)
# encountered an error scipy not found installed the package
```

```python
#A heatmap is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions
corr_matrix=df.corr(method ='pearson',numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized=df.copy()
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
```

```python
corr_matrix=df_numerized.corr(method ='pearson',numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
